```python
import requests, os
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import time
```

```python
path = "landscape-4.jpeg"
```

```python
plt.imshow(pltimg.imread(path))
```

```python
url = 'http://34.76.227.221/translate'
files = {'file': open(path, 'rb')}
start_time = time.time()
r = requests.post(url, files=files, headers={'Host': 'app.default.svc.cluster.local'}, verify=False)
print("{} Sekunden".format(time.time()-start_time))
```

```python
print(r)
```

```python
open('landscape-mama-vangogh.png', 'wb').write(r.content)
```
